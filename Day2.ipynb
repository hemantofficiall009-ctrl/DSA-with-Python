```python
# Find largest number in an array
arr=[-1,-2,-1.5,0,-0.1]
largest=float('-inf')
seclargest=float('-inf')

for i in arr:

  if i > largest:
    seclargest=largest
    largest=i
  elif i<largest and i>seclargest:
     seclargest=i
print(largest)
print(seclargest)
```

```python
# Move All Zeros to End
arr = [0, 0,2]
for i in arr:
  if i==0:
    arr.remove(i)
    arr.append(i)
print(arr)
```

```python
# Remove duplicates (in-place)
arr = [1,2,2,3,3,4]
j=0
for i in range (1,len(arr)):
   if arr[i] != arr[j]:
     j+=1
     arr[j]=arr[i]
print(arr[:j+1])
```

```python
# Remove a Specific Element
arr = [3, 2, 2, 3.1,3]
val=3
j=0
for i in range(len(arr)):
  if arr[i]!=val:
    arr[j]=arr[i]
    j+=1
a=arr[:j]
print(a)
```

```python
# Check if a string is a Palindrome (reads the same forward and backward, like "racecar").
s=input("Enter any string to check it ia a palindrome")
reversed_s = s[::-1]
for i in range(len(s)-1):
  if s==reversed_s:
    print("This is a palindrome string")
    break
  else :
    print("This is not a plindrome")
    break
```
