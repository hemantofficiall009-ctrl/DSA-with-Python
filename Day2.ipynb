```python
# Check if a string is a Palindrome (reads the same forward and backward, like "racecar").
s=input("Enter any string to check it ia a palindrome")
reversed_s = s[::-1]
for i in range(len(s)-1):
  if s.lower()==reversed_s.lower():
    print("This is a palindrome string")
    break
  else :
    print("This is not a plindrome")
    break
```

```python
# Find the Middle of a List
arr = [1, 2, 3, 4, 5, 6]
slow = 0
fast = 0

# Loop as long as 'fast' can still move 2 steps forward
while fast < len(arr) and fast + 1 < len(arr):
     slow+=1
     fast+=2

print("The middle element is:", arr[slow])
```

```python
# You are given a sorted list of numbers and a target sum.
#  Find the indices of 000000000000000the two numbers that add up to that target.
# Input: nums = [2, 7, 11, 15], target = 9Output: 0, 1 (because 2 + 7 = 9)
nums=[-3,-3.3,2,7,11,15]
target=12
left=0
right=len(nums)-1
while left<right:
  sum=nums[left]+nums[right]
  if sum==target:
    print(f"I found the indices these are {left},{right}")
    break
  elif sum<target:
    left+=1
  elif sum>target:
    right-=1
```
