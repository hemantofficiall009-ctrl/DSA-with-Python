```python
# Find largest number in an array
arr=[-1,-2,-1.5,0,-0.1]
largest=float('-inf')
seclargest=float('-inf')

for i in arr:

  if i > largest:
    seclargest=largest
    largest=i
  elif i<largest and i>seclargest:
     seclargest=i
print(largest)
print(seclargest)
```
