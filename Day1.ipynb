```python
# Find largest number in an array
arr=[-1,-2,-1.5,0,-0.1]
largest=float('-inf')
seclargest=float('-inf')

for i in arr:

  if i > largest:
    seclargest=largest
    largest=i
  elif i<largest and i>seclargest:
     seclargest=i
print(largest)
print(seclargest)
```

```python
# Move All Zeros to End
arr = [0, 0,2]
for i in arr:
  if i==0:
    arr.remove(i)
    arr.append(i)
print(arr)
```

```python
# Remove duplicates (in-place)
arr = [1,2,2,3,3,4]
j=0
for i in range (1,len(arr)):

  if arr[j] != arr[i]:
    j+=1
    arr[j] = arr[i]
print(arr[:j+1])
```

```python
arr = [0,0,1,1,1,2,2,3]
j=0
for i in range(len(arr)):
  if arr[j]!=arr[i]:
    j+=1
    arr[j]=arr[i]
print(arr[:j+1])
```

```python
# Allow only 2 duplicates
arr = [1,1,1,2,2,2,2,2,2,2,3,3,3,3,3,3,3,3,3]
j=0
for i in range (1,len(arr)):
    if arr[j-2] !=arr[i]:
        j+=1
        arr[j]=arr[i]
print(arr[:j+1])

```
