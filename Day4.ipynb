## 1. Variables

In Python, a variable is a named location used to store data in the memory. It's like a container that holds a value. When you create a variable, you reserve a space in memory for that value. The value stored in a variable can be changed during program execution.

### Naming Conventions:
*   Variable names must start with a letter or an underscore `_`.
*   They cannot start with a number.
*   They can only contain alpha-numeric characters and underscores (A-z, 0-9, and _).
*   Variable names are case-sensitive (`age`, `Age`, and `AGE` are three different variables).
*   It's good practice to use meaningful names.

### Assignment:
To assign a value to a variable, you use the assignment operator (`=`).

```python
# Assigning different types of values to variables
my_integer = 10
my_float = 20.5
my_string = "Hello, Python!"
is_active = True

# You can also change the value of a variable
my_integer = 15 # Now my_integer holds a new value

# Print the values of the variables to see what they hold
print(f"Integer variable: {my_integer}")
print(f"Float variable: {my_float}")
print(f"String variable: {my_string}")
print(f"Boolean variable: {is_active}")

# You can also assign the result of an expression to a variable
sum_of_numbers = my_integer + my_float
print(f"Sum of numbers: {sum_of_numbers}")
```

## 2. Data Types

In Python, every value has a data type. Understanding these types is essential for performing correct operations and handling data effectively. Python has several built-in data types, including:

*   **Numeric Types**: `int` (integers), `float` (floating-point numbers), `complex` (complex numbers).
*   **Sequence Types**: `str` (strings), `list`, `tuple`.
*   **Mapping Type**: `dict` (dictionaries).
*   **Set Types**: `set`, `frozenset`.
*   **Boolean Type**: `bool` (True/False).

Let's look at examples of some common data types.

```python
# Integer (int)
age = 30
print(f"Age: {age}, Type: {type(age)}")

# Float (float)
price = 19.99
print(f"Price: {price}, Type: {type(price)}")

# String (str)
name = "Alice"
message = 'Hello World'
print(f"Name: {name}, Type: {type(name)}")
print(f"Message: {message}, Type: {type(message)}")

# Boolean (bool)
is_student = True
has_license = False
print(f"Is Student: {is_student}, Type: {type(is_student)}")
print(f"Has License: {has_license}, Type: {type(has_license)}")

# List (list) - ordered, changeable, allows duplicates
numbers = [1, 2, 3, 4, 5]
mixed_list = ["apple", 10, True]
print(f"Numbers: {numbers}, Type: {type(numbers)}")
print(f"Mixed List: {mixed_list}, Type: {type(mixed_list)}")

# Tuple (tuple) - ordered, unchangeable, allows duplicates
coordinates = (10.0, 20.0)
print(f"Coordinates: {coordinates}, Type: {type(coordinates)}")

# Dictionary (dict) - unordered, changeable, no duplicate keys
person = {"name": "Bob", "age": 25, "city": "New York"}
print(f"Person: {person}, Type: {type(person)}")

# Set (set) - unordered, unchangeable*, no duplicate members
colors = {"red", "green", "blue"}
print(f"Colors: {colors}, Type: {type(colors)}")

```

## 3. Operators

Operators are special symbols that perform operations on one or more values (operands). These operations can range from simple arithmetic to complex logical comparisons. Understanding operators is crucial for writing any kind of computational logic.

Python supports several types of operators:

*   **Arithmetic Operators**: Perform mathematical calculations.
*   **Comparison Operators**: Compare two values and return a boolean result (`True` or `False`).
*   **Logical Operators**: Combine conditional statements.
*   **Assignment Operators**: Assign values to variables.
*   **Identity Operators**: Compare the memory locations of two objects.
*   **Membership Operators**: Test if a sequence contains a specific value.

```python
# --- Arithmetic Operators ---
x = 10
y = 3

print(f"x + y = {x + y}")  # Addition
print(f"x - y = {x - y}")  # Subtraction
print(f"x * y = {x * y}")  # Multiplication
print(f"x / y = {x / y}")  # Division (returns float)
print(f"x // y = {x // y}") # Floor Division (returns integer)
print(f"x % y = {x % y}")  # Modulus (remainder)
print(f"x ** y = {x ** y}") # Exponentiation

print("\n---") # Separator

# --- Comparison Operators ---
a = 5
b = 10

print(f"a == b : {a == b}") # Equal to
print(f"a != b : {a != b}") # Not equal to
print(f"a > b : {a > b}")  # Greater than
print(f"a < b : {a < b}")  # Less than
print(f"a >= b : {a >= b}") # Greater than or equal to
print(f"a <= b : {a <= b}") # Less than or equal to

print("\n---") # Separator

# --- Logical Operators ---
# Used to combine conditional statements
is_sunny = True
is_warm = False

print(f"is_sunny and is_warm : {is_sunny and is_warm}") # Logical AND
print(f"is_sunny or is_warm : {is_sunny or is_warm}")  # Logical OR
print(f"not is_sunny : {not is_sunny}")          # Logical NOT

print("\n---") # Separator

# --- Assignment Operators ---
c = 10
print(f"Initial c: {c}")
c += 5  # Equivalent to c = c + 5
print(f"c after c += 5: {c}")
c -= 3  # Equivalent to c = c - 3
print(f"c after c -= 3: {c}")

print("\n---") # Separator

# --- Identity Operators ---
# 'is' and 'is not' compare if two variables point to the same object in memory
list1 = [1, 2, 3]
list2 = [1, 2, 3]
list3 = list1

print(f"list1 is list2 : {list1 is list2}") # False, different objects
print(f"list1 == list2 : {list1 == list2}") # True, same values
print(f"list1 is list3 : {list1 is list3}") # True, list3 points to list1

print("\n---") # Separator

# --- Membership Operators ---
# 'in' and 'not in' test for membership in a sequence
my_list = [10, 20, 30, 40]
print(f"20 in my_list : {20 in my_list}")
print(f"50 not in my_list : {50 not in my_list}")

my_string = "Python Programming"
print(f"'Program' in my_string : {'Program' in my_string}")
```

```python
my_set = {"apple", "banana", "cherry"}

# Test for membership
print(f"Is 'banana' in my_set? {'banana' in my_set}")
print(f"Is 'grape' in my_set? {'grape' in my_set}")
```
